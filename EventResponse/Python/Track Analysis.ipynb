## Event Response Track Analysis

Purpose of these analysis is find out simulated events whose tracks are intersecting with upcoming event based NHC Data.

Python script works as following:
1. Read input data which include Polygon shape file (Source: `NHC`), Line shape file (Source: `NHC`), Events track shape file and US map shape file
2. Findout `num_points` from Line shape file which are equidistance
3. Using those points, create circles which are within polygon using polygon shape file
4. Findout events which are passing through all circles

### Input Files:
1. Simulated Event Tracks provided by AIR
2. NHC Data - Event Advisories

### Output:
1. Filtered events
